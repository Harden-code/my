```python
import sympy as sp
```

```python
x,y=sp.symbols('x,y')
sp.Integral(x**2*y,(y,0,sp.sqrt(1-x**2)),(x,0,1)).doit()
```

```python
sp.simplify(sp.Integral(sp.E**(x+y),(y,0,1),(x,0,1)).doit())
```
