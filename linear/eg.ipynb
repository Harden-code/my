```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


theta=np.linspace(0,2*np.pi,30)
x=np.cos(theta)
y=np.sin(theta)

fig,*ax=plt.subplots(1,2,figsize=(6,6))
org_xy=np.full_like(x,0)
ax[0][0].quiver(org_xy, org_xy, x, y, 
          pivot='mid',      # 锚点设为中间，更整齐
          scale=1,          # 【核心】调大这个数，箭头就会变小
          scale_units='inches', # 配合 scale 使用，保证缩放一致
          width=0.005,      # 调小这个数，箭杆变细
          headwidth=3.5,    # 箭头头部宽度
          headlength=4.5,   # 箭头头部长度
          color='#1f77b4')

ax[0][0].set_aspect('equal')  # 必须设置，否则圆会变成椭圆，箭头也会变形
ax[0][0].set_xlim(-1.5, 1.5)
ax[0][0].set_ylim(-1.5, 1.5)


A=np.array([[1.25,-0.75],[-0.75,1.25]])
B=A@np.array([x,y])
ax[0][0].quiver(org_xy, org_xy, B[0,], B[1,], 
          pivot='mid',      # 锚点设为中间，更整齐
          scale=1,          # 【核心】调大这个数，箭头就会变小
          scale_units='inches', # 配合 scale 使用，保证缩放一致
          width=0.005,      # 调小这个数，箭杆变细
          headwidth=3.5,    # 箭头头部宽度
          headlength=4.5,   # 箭头头部长度
          alpha=0.4,
          color="#b41fa0")


ax[0][1].quiver(org_xy, org_xy, B[0,], B[1,], 
          pivot='mid',      # 锚点设为中间，更整齐
          scale=1,          # 【核心】调大这个数，箭头就会变小
          scale_units='inches', # 配合 scale 使用，保证缩放一致
          width=0.005,      # 调小这个数，箭杆变细
          headwidth=1.5,    # 箭头头部宽度
          headlength=4.5,   # 箭头头部长度
          alpha=0.4,
          color='#1f77b4')

ax[0][1].set_aspect('equal')  # 必须设置，否则圆会变成椭圆，箭头也会变形
ax[0][1].set_xlim(-1.5, 1.5)
ax[0][1].set_ylim(-1.5, 1.5)
```
