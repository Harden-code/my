```python
import sympy as sp
x,y,n,k=sp.symbols('x,y,n,k',integer=True)
```

```python
sp.expand((x+y)**4)
```

```python
sp.Sum(sp.binomial(4,k)*x**(4-k)*y**k,(k,0,4)).doit()
```

```python
sp.Sum(sp.binomial(4,k)*x**(4-k)*y**k,(k,0,4)).doit().subs({x:1,y:-1})
```

### 容斥定理

注意容斥定理与二项式展开的,交集的符号与二项式展开符号

利用$(x+y)^n$ x=-1,y=1的形式构造出容斥的形式
