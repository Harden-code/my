### 独立性
在看待$pr(A \cap  B)= pr(A)*pr(B)$,独立表示事件A发生的概率不会影响事件B发生的概率,$pr (A \cap B)$得从发生这两件事的结果集来看待,A和B同时发生的概率

```python
import sympy as sp
n=sp.symbols('n',postive=True,integer=True)
```

```python
deck=[(i,j) for j in range(1,5) for i in range(1,14)]

count=0
all=1000000
import random
for _ in range(all):
    hand=random.sample(deck,5)
    samebit={card[0] for card in hand} 
    if len(samebit)==5:
        count+=1

count/all
```
