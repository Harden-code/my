```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x,u,o=sp.symbols('x,u,o',real=True,positive=True)
fX=1/(sp.sqrt(2*sp.pi*o**2))*sp.E**-(((x-u)**2)/(2*o**2))
fX
```

```python
sp.Integral(fX,(x,-sp.oo,sp.oo)).doit()
```

```python
# 均值
sp.Integral(x*fX,(x,-sp.oo,sp.oo)).doit()
```

```python

f=sp.E**-(((x-u)**2)/(2*o**2))
sp.Integral(f,(x,-sp.oo,sp.oo)).doit()
```

```python
f_0_1=fX.subs({o:2,u:1})
f_0_1
```

```python
f_0_1_f=sp.lambdify(x,f_0_1)

idx=np.linspace(-4,4,101)
plt.plot(idx,f_0_1_f(idx))
```
