```python
import sympy as sp
```

### 符号假设
合理使用符号假设,它默认的假设是虚数

Ratianal精确表示有理数,比如1/3而不是构造代数分式

```python
# combsimp简化组合的函数,如果符号设置为integer简化就不会变成Gamma函数
```

```python
n,k=sp.symbols('n,k')
expr=sp.binomial(n,k)-sp.binomial(n,k-1)
sp.combsimp(expr)
```

```python
n,k=sp.symbols('n,k',integer=True)
expr=sp.binomial(n,k)-sp.binomial(n,k-1)
sp.combsimp(expr)
```

```python
# 简化思路 重写rewrite表达方式
expr_fact=expr.rewrite(sp.factorial)
# 在进行简化
sp.simplify(expr_fact)
```

### 简化

多项式相关

指数对数相关

组合数相关

expr.rewrite强制将一种函数转换为另外一种

expr.factor_terms 提取表达式所有项的公共因数

```python
x,y=sp.symbols('x,y')
expr=(x+y)**2
```

```python
# simplity 简化表达式,有些时候不会做任何操作,计算开销大
```

```python
# expand 展开
sp.expand(expr)
```

```python
# factor 因式分解,提取公因式,将其化为乘积
sp.factor(x**2-y**2)
```

```python
# 合并同类项,按照指定变量var来收集
sp.collect(x*y + x - 3 + 2*y- y + x, x)
```

```python
sp.collect(x*y+x**3+x**2-y-1, (x,y))
```

```python
# cancel 通分,把p/q标准形式消除分子分母的公因式
sp.cancel((x**2-1)/(x-1))
```

```python
# apart 部分分式展开;微分求积分利器
sp.apart(1/(x**2-x))
```

```python
# collect 合并相同次幂的系数;想要提出直接使用因式分解
sp.collect(x**3+x**2+x,x)
```

```python
sp.collect(x**2+y*x**2+x,x)
```

```python
#三角函数展开
expr=sp.sin(x+y)
expr_expnd=sp.expand_trig(expr)
expr_expnd
```

```python
sp.expand_func
```

```python
# 三角函数化简
sp.trigsimp(expr_expnd)
sp.trigamma
```

```python
#指数 注意假设

a,b=sp.symbols('a,b',integer=True)
x,y=sp.symbols('x,y',positive=True,integer=True)
expr=x**a*x**b
# 指数合并
expr_simp=sp.powsimp(expr)
expr_simp
```

```python
# 指数展开
sp.expand_power_exp(expr_simp)
```

```python
# 对数展开
expr=sp.log(x*y)
expr_expa=sp.expand_log(expr)
expr_expa
```

```python
# 对数合并
sp.logcombine(expr_expa)
```

```python
# 排列
n,k=sp.symbols('n,k',integer=True,positive=True)
ff=sp.ff(n,k)
ff.rewrite(sp.factorial)
```
