```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
np.random.seed(10)
data=np.random.normal(0,0.5,1000)

cumsum=np.cumsum(data)

plt.plot(cumsum, label='data')
```

```python
x = sp.symbols('x')

expr=(x**2-4)/(x-2)

sp.pprint(expr)

sp.plot(expr, (x, -10, 10), title='Plot of the expression', xlabel='x', ylabel='f(x)', show=True)
```
