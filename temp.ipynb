```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0.1, 8, 100)
y2=np.log(x)
y1=np.exp(y2)

fig, ax = plt.subplots()
ax.plot(y2, y1, label='exp(x)')
ax.plot(x, y2, label='log(x)')
ax.legend()
ax.set_aspect('equal')
ax.set_xlim(0, np.max(x)*1.1)
ax.set_ylim(0, np.max(x)*1.1)
ax.set_xticks(np.arange(0, np.max(x)*1.1, 1))
ax.set_yticks(np.arange(0, np.max(x)*1.1, 1))
ax.grid(True)
```

```python
from sympy import *
x,n,i=symbols('x,n,i')
f=x**2
fdiff=f.diff(x)
sup=4
sub=1
idx=(sup-sub)/n
xi=sub+idx*i
fsum=2*xi*idx
Integral(fdiff,(x,sub,sup)).doit(),Sum(fsum,(i,1,n)).evalf(subs={'n':1000})
```

```python
from sympy import *

x, n, i = symbols("x, n, i")
a=1
b=4
delta=(b-a)/n 
deltaX=a+(b-a)/n*i
f=(1/deltaX)*delta
# Sum(f,(i,a,n))
Sum(f,(i,a,n)).evalf(subs={n:1000})
ln(4).evalf(),Sum(f,(i,a,n)).evalf(subs={n:1000})

```
