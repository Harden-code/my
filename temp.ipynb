```python
from sympy import *
x,n,i=symbols('x,n,i')
f=x**2
fdiff=f.diff(x)
sup=4
sub=1
idx=(sup-sub)/n
xi=sub+idx*i
fsum=2*xi*idx
Integral(fdiff,(x,sub,sup)).doit(),Sum(fsum,(i,1,n)).evalf(subs={'n':1000})
```

```python
x,n=sp.symbols('x,n')
```

$\sum_{a=1}^n $ 名字叫西格玛,累加符号

用来解决什么问题了?

就是为了偷懒,假想一种简单的场景,你想表示n个2相加,直观方式2+2+2...
但是这样表示太繁琐.

所以数学家定义一个通用的方式$\sum_{a=1}^n 2$ 下标表示第一个元素,上标表示
第n个元素

```python
sp.Sum(2,(x,1,n))
```

```python
sp.Sum(2,(x,1,n)).doit()
```

```python
sp.Sum(x,(x,1,n))
```

```python
sp.Sum(x,(x,1,n)).doit()
```
