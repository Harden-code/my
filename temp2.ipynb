```python
import sympy as sp
p=sp.symbols('p')
n=sp.symbols('n',real=True,positive=True)

expr=p*(1-p)**(n-1)
expr
```

```python
r=p/(1-(1-p))
r
```

```python
sum_expr=sp.Sum(expr,(n,0,n))
sum_expr # fX =1/1-p
```

```python
dFx=sum_expr.diff(p)
dFx
```

```python
sp.simplify(sp.cancel(dFx.args[0]))
```

```python
lc=(r.diff(p)*p*(1-p)*(n))/(-n*p+1)
sp.simplify(lc)
```

```python
import sympy as sp
x,n=sp.symbols('x,n',positive=True)
sp.Integral(x**n*sp.E**-x,(x,0,x)).doit()
```

```python
fx=x**n*sp.E**-x
sp.init_printing(pretty_print=True)

for _ in range(1,4):
    sp.pprint(sp.Integral(fx.subs(n,_),(x,0,sp.oo)).doit())
```

```python
sp.Integral(fx.subs(n,1),(x,0,sp.oo)).doit()
```
