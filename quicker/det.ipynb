### 行列式
行列式的含义线性放方程解的分母

```python
import sympy as sp
```

```python
x,y,z=sp.symbols('x,y,z',real=True)
a,b,c,d,e,f,g,h,i,j=sp.symbols('a:j',integer=True)
```

```python
A=sp.Matrix([[a,b,c],[d,e,f],[g,h,i]])
sp.det(A)
```

```python
f1=a*x+b*y+c*z-j
f2=d*x+e*y+f*z-j
f3=g*x+h*y+i*z-j

sp.linsolve([f1,f2,f3],[x,y,z])
```
