```python
# y=x^2 Âú®x=3‰∏ãÁöÑÈù¢ÁßØ
x,n=sp.symbols('x,n')
A_n=(3/n)**3
fig,ax=plt.subplots()
for i in range(5,101):
    y=A_n.subs(n,i) *sp.Sum((n-1)**2,(n,1,n)).subs(n,i).evalf()
    ax.scatter(i,y,c='b',alpha=0.9,s=0.05)
y
```
