### 连续性
$\lim_{x \to a} f(x)=f(a)$

特别例子:
$\begin{cases}
xsin(1/x)   &当x !=0\\
0   &当x=0 \\
\end{cases}$

这个xsin(1/x)本身在0点没意义,因为这里补了0所以该分段函数是连续的

连续性的一个作用:
1. 在求函数在某点的极限时,可以通过连续的性质进行关联.
2. 介值定理:f(x)在[a,b]之间连续,并且f(a)<0且f(b)>0,那么该区间[a,b]上至少有一点c,使得f(c)=0.反之f(a)>0且f(b)<0也成立
> trick 奇数次的多项式的解,始终有大于0和小于0的值

### 可导性(光滑性)
严格定义:设函数y=f(x)在$x_0$某个领域内有定义,当自变量x在$x_0$处取得增量$\Delta x$(点$x_0 + \Delta x$仍在领域内),相应的,因变量$\Delta y= f(x_0 + \Delta x)-f(x_0)$.如果当$\Delta y 与 \Delta x$之比当$\Delta x \to 0$时极限存在,那么$y=f(x_0)$处可导

$f(x)^\prime=\lim_{\Delta x \to 0} \frac{\Delta y}{\Delta x}$

> 比如 $\left \vert x \right \vert$ 在x=0处,由于$0^-和0^+$左右极限不等,极限不存在所有没有导数


线性近似:$f(x)=x^2$,求2.001就可以直接通过$f(x)+\Delta y, \Delta y=f(2)^\prime * \Delta 0.001$

### 连续性与可导性的关系
如果一个函数可导则它一定是连续的

> 注意连续的性质在该点连续,则极限等于该点的值.而导数$f(x)^\prime = lim_{h \to 0} \frac{f(x+h)-f(x)}{h} * \lim_{h \to 0}h$ 左边等于0,因此$lim_{h \to 0} f(x+h)=f(x)$ 

```python
from sympy import *
x,y=symbols('x,y')
```

```python
f=x**2
f1=diff(f)
# 线性估计
f.subs(x,2)+f1.subs(x,2)*0.0004,f.subs(x,2.0004)
```

### 多项式求导式证明

#### 乘法法则
uv看成一个长方形面积扩张$(u+\Delta u) * \Delta v + (v+\Delta v) * \Delta u$

```python
f1,f2=symbols('f1,f2',cls=Function)
fun_multiply=f1(x)*f2(x)
diff(fun_multiply)
```

#### 链式法则
假设u=f(x),y=g(u).x变化会导致u的变化,而u的变化导致v得变化.因此x变化会导致u拉长$f(x)^\prime$倍,然后乘上原本$g(x)^\prime$倍

$\Delta u=f(x)^\prime * \Delta x  $

$  \Delta y=g(u)^\prime * \Delta u$ 

$\frac{\Delta y}{\Delta x} = g(u)^\prime * f(x)^\prime  $


```python
fun_channel=f1(f2(x))
diff(fun_channel)
```
