
### sympy使用隐函数求导注意
1. 使用diff求导
2. solve求解$y^\prime$

> 坑点:在设置变量时如果不断言sympy会默认为实数,而隐函数是y是一个函数

```python
from sympy import *
x=symbols('x')
y=symbols('y',cls=Function)

expr=x**2+y(x)**2-4

expr_diff=diff(expr,x)
pprint(expr_diff)

solve(expr_diff,diff(y(x),x))
```

```python
# 计算隐函数微分
idiff(expr,y(x),x)
```

```python
# 表示是一个导数不立即求值,用于构建方程或数学打印,用doit直接展开
Derivative(x**2+x,x)
```
