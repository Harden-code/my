```python
import numpy as np
import sympy as sp
a,b,c,d,e,f,g,h,i,j=sp.symbols('a,b,c,d,e,f,g,h,i,j')

A_1=sp.Matrix([[a,b],[c,d]])
A_2=sp.Matrix([[a,b,c],[e,f,g],[h,i,j]])
```

特征值的性质
1. $\det(A)=\lambda_1*\lambda_2...$
2. $tr(A)=\lambda_1+\lambda_2..=a_{11}+a_{22}..$



```python
# 
print("2x2 det:",end=' ');sp.pprint(A_1.det())
print("2x2 trace:",end=' ');sp.pprint(A_1.trace())
print("3x3 det :",end=' ');sp.pprint(A_2.det())
print("3x3 trace:",end=' ');sp.pprint(A_2.trace())
```

```python
la_1,la_2=A_1.eigenvals()
sp.simplify(la_1*la_2-(A_1.det()))
sp.expand(la_1*la_2)
```

```python
la3_1,la3_2,la3_3=A_2.eigenvals()
la3_1*la3_2*la3_3
sp.simplify(la3_1*la3_2*la3_3-(A_2.det()))
```

```python
sp.simplify(la_1+la_2-A_1.trace())
```

```python
# LU 
A=sp.Matrix([[1,2],[2,1]])
L,U,C=A.LUdecomposition()
print(A.eigenvals())
sp.pprint(L)
sp.pprint(U)
```

```python
# 对角化
P,D=A.diagonalize()

P*D*P**-1
# D 对角,也就是两个特征值
D
```
