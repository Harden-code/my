### 一切的来源
#### 小数情况
$\lim_{x \to 0} \frac{sin(x)}{x}=1$

$\lim_{x \to 0}cos(x)=0$

$\lim_{x \to 0} \frac{tan(x)}{x}=\lim_{x \to 0} \frac{\frac{sin(x)}{cos(x)}}{x}=\lim_{x \to 0} \frac{sin(x)}{x}*\frac{1}{cos(x)}=1$

特别注意

$\lim_{x \to 0} \frac{1-cos(x)}{x}=\lim_{x \to 0} \frac{1-cos^2(x)}{x*(1+cos(x))}=\lim_{x \to 0} sin(x)*\frac{sin(x)}{x}*\frac{1}{(1+cos(x))}=0$

$\lim_{x \to 0} \frac{1-cos(x)}{x^2}=\lim_{x \to 0} \frac{1-cos^2(x)}{x^2*(1-cos(x))}=\lim_{x \to 0} \frac{1-cos^2(x)}{x^2*(1+cos(x))}=\lim_{x \to 0} \frac{sin^2(x)}{x^2*(1+cos(x))}=\lim_{x \to 0} \frac{sin^2(x)}{x^2} *\frac{1}{(1+cos(x))}=\frac{1}{2}$

#### 大数情况

$\lim_{x \to \infty} \frac{sin(any)}{x^q}=0$

证明主要用三明治定理来证明(-1<=sin(x)<=1)


### 变体情况
$\lim_{x \to \frac{\pi}{2}}\frac{cos(x)}{x-\frac{\pi}{2}} 令 t=x- \frac{\pi}{2} 则 x=t+\frac{\pi}{2} ,原式等于 \lim_{t \to 0} \frac{cos(t+ \frac{\pi}{2})}{t} 然后在三角转换$


$\lim_{t \to 0} {sin(\frac{1}{x})}当x=1/\pi,2/\pi函数就一直在跳动$

### $\lim_{x \to 0} \frac{sin(x)}{x}=1$
利用内切三角形<=扇形<=外切三角形面积(三明治原则)证明


### 三角函数导数
导数主要通上面极限来进行计算

```python
import sympy as sp
x=sp.symbols('x')
# y=-16a
expr=3*sp.sin(4*x)
# -16y=sp.diff(expr,x,2) 加速度
sp.pprint(sp.diff(expr,x,2))
```
