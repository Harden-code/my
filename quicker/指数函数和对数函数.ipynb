### 指数函数和对数函数
目的将乘法变成加法来简化计算

$M=b^m N=b^n$
- revise rule

    $m= \log_{b}M $
    
    $M=b^m=b^{\log_{b}M}$
- product rule

    $MN=b^m*b^n取对数\log_{b}{MN}=\log_{b}{b^{m+n}}=m+n=\log_{b}M + \log_{b} N$

- quotient rule

    $\frac{M}{N}取对数\log_{b}{b^{m-n}}=m-n=\log_{b}{M}-\log_{b}{N}$

- power rule

    $M^2取对数\log_{b}b^2m=m+m=2\log_{b}m$

- change of base

    $\log_{b}M=\frac{\log_{n}M}{\log_{n}b}$

    prove:

    $y=\log_{b}M利用一对一关系取以c为底M为值的对象数\log_{c}{M}=\log_{c}b^y,在利用power法则 \log_{c}{M}=y*\log_{c}b化简 y=
\frac{\log_{c}{M}}{\log_{c}b}$

```python
import numpy as np
import matplotlib.pyplot as plt

x1=np.arange(0,10,1)
x2=np.arange(0,20,1)
x3=np.arange(0,30,1)
y1=(1+0.12)**x1
y2=(1+0.12)**x2
y3=(1+0.12)**x3

fig,*ax=plt.subplots(3,1,layout='constrained',figsize=(10,8))
ax[0][0].plot(x1,y1)
ax[0][0].set_aspect('equal')
ax[0][0].set_xlim([0,40])
ax[0][0].set_ylim([0,30])
ax[0][0].grid()

ax[0][1].plot(x2,y2)
ax[0][1].set_aspect('equal')
ax[0][1].set_xlim([0,40])
ax[0][1].set_ylim([0,30])
ax[0][1].grid()

ax[0][2].plot(x3,y3)
ax[0][2].set_aspect('equal')
ax[0][2].set_xlim([0,40])
ax[0][2].set_ylim([0,30])

ax[0][2].grid()
```
