### 矩阵乘法
矩阵乘法,满足的前提线性映射:f(x+y)=f(x)+f(y),cf(x)=f(cx)

本质符合运算f(x) g(x)  f(g(x))
f(x)与g(x)符合满足未知数的个数都是相同,在矩阵乘法中反映的就是:两个矩阵相乘第一个矩阵列数要与第二矩阵行数相同.第一个矩阵的列代表它未知数个数,第二行数代表它未知数个数

```python
import sympy as sp
import matplotlib.pyplot as plt
sp.init_printing()
a,b,c,d=sp.symbols('a:d')
e,f,g,h=sp.symbols('e:h')
x,y=sp.symbols('x,y')
x1,x2,y1,y2=sp.symbols('x1:3,y1:3')
```

```python
A=sp.Matrix([[a,b],[c,d]])
B=sp.Matrix([[e,f],[g,h]])
x=sp.Matrix([x1,x2]).T
x,A,x@A
```

```python
f1=a*x1+b*y1
f2=c*x1+d*y1

g1=e*x+f*y
g2=g*x+h*y

sp.collect(sp.expand(f1.subs([(x1,g1),(y1,g2)])),(x,y))
```
