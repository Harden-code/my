```python
#习题1.5.3 我们已经证明了,当房间里有人23时,至少有两人生日在同一天的概率为50%.平均来说,需要多少人才能使至少有两人生日在同一天？
# 如果用通用公式会log(1)=0
cur=1
for i in range(2,100):
    newfact=(365-(i-1))/365
    cur*=newfact
    same_day=1-cur
    if same_day>=1:
        print(i)
    if i>90:
        print(i,same_day)
```
