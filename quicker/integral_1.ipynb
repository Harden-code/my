```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import RegularPolygon,Circle
import sympy as sp
```

```python
fig,ax =plt.subplots()
n=10
circle=Circle(xy=(0,0),radius=1,alpha=0.8)

inner=RegularPolygon(xy=(0,0),radius=1,numVertices=n)
r=np.divide(1,np.cos(np.pi/n))
out=RegularPolygon(xy=(0,0),numVertices=n,radius=r,alpha=0.4,facecolor='red')

ax.add_patch(inner)
ax.add_patch(circle)
ax.add_patch(out)


ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.set_aspect('equal')
```

```python
i=sp.symbols('i')
# 注意(idx,start,end)格式
# 累加
sp.Sum(i,(i,1,5))

# 常数累加
sp.Sum(2,(i,1,5))

```

```python
x=sp.symbols('x')
integral_sin=sp.Integral(3*sp.sin(x),(x,0,x))
# sp.diff(integral_sin)
integral_sin
```
