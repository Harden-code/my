```python
import numpy as np
import matplotlib.pyplot as plt

data=np.genfromtxt('ex2data1.txt',delimiter=',')

X=data[:,0]
y=data[:,-1]
```

```python
F=y==0
F_X=X[F]
T_X=X[~F]

fig,ax=plt.subplots()
F_Y=np.zeros_like(F_X)
T_Y=np.ones_like(T_X)
ax.scatter(F_X,F_Y,marker='x')
ax.scatter(T_X,T_Y,marker='.')
ax.plot(X,0.0109*X-0.0005)
```

```python

def err(y,y_hat):
    e=y_hat-y
    return np.dot(e,e)/len(y)

def sigmoid(x,w,b):
    z=w*x+b
    return 1/(1+np.exp(-z))

def df(X,y,w,b):
    dw=0
    db=0
    M=X.shape[0]
    errs=np.zeros(M) #每列的差值
    for i in range(M):
        y_hat=sigmoid(X[i],w,b)
        err=y_hat-y[i]
        dw+=X[i]*err
        db+=err
        errs[i]=err

    return dw/M,db/M,errs

def fit(X,y,count,alpha):
    w=0
    b=0
    result=[]
    for _ in range(count):
        dw,db,y_hat=df(X,y,w,b)
        w=w-alpha*dw
        b=b-alpha*db
        e=err(y,y_hat)
        print(f"迭代 {_+1:03d}次 | w: {w:8.4f} | b: {b:8.4f} | Error: {e:10.4f}") 
        result.append((w,b,e))

    
```

```python
fit(X,y,100,0.0001)
```
