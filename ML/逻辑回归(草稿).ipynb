```python
import sympy as sp
x,a,b=sp.symbols('x,a,b')
z=a*x+b
sigmoid=1/(1+sp.exp(-z))
```

```python
sp.simplify(sigmoid*(1-sigmoid)-sp.diff(sigmoid,x))
```

```python
y,y_hat=sp.symbols('y,y_hat')

expr=-(y*sp.ln(y_hat)+(1-y)*sp.ln(1-y_hat))
expr
```

```python
L=expr.subs({y_hat:sigmoid})
L
```

```python
dL_da=sp.diff(L,a)
```

```python
simplified_dL_da = sp.simplify(dL_da)
simplified_dL_da.subs(sigmoid,y_hat)
```

```python
# 1. 定义我们需要的所有的符号变量
# x: 特征, y: 真实标签, a: 权重(相当于w), b: 偏置
x, y, a, b = sp.symbols('x y a b')

# 2. 第一层：定义线性组合 z
z = a * x + b

# 3. 第二层：定义 Sigmoid 函数预测值 y_hat 
# 注意：Sympy中用 sp.exp() 表示 e 的指数
y_hat = 1 / (1 + sp.exp(-z))

# 4. 第三层：定义交叉熵损失函数 L
L = -(y * sp.log(y_hat) + (1 - y) * sp.log(1 - y_hat))

# --- 下面开始见证奇迹 ---

# 5. 对参数 a (也就是权重w) 求偏导数
dL_da = sp.diff(L, a)

print("未化简的原始导数（会非常长且复杂）：")
print(dL_da)
print("-" * 50)

# 6. 使用 sp.simplify() 让 Sympy 进行代数化简！
# 此时 Sympy 会把它合并成一个大分数，消除负指数
simplified_dL_da = sp.simplify(dL_da)

print("经过初步化简后的导数（合并为了大分式）：")
print(simplified_dL_da)

# 7. 终极奥义：引导 Sympy 按照人类的习惯替换出 y_hat
# 因为化简后的式子里全都是正指数 e^(ax+b)，也就是 e^z
# 根据 y_hat = e^z / (1 + e^z)，可以反解出： e^z = y_hat / (1 - y_hat)
y_hat_symbol = sp.Symbol('y_hat')
e_z = sp.exp(a * x + b)  # 提取出式子里的指数项

# 我们告诉 Sympy：把所有的 e^(ax+b) 都替换成 y_hat / (1 - y_hat)，然后再化简一次！
final_beautiful_result = sp.simplify(simplified_dL_da.subs(e_z, y_hat_symbol / (1 - y_hat_symbol)))
final_beautiful_result
```
