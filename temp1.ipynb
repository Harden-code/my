```python
# 使用函数符合+线性转换视角
x1,x2,y1,y2=symbols('x1 x2 y1 y2')
a,b,c,d=symbols('a:d')
p,q,r,s=symbols('p:s')

xx1=a*x1+b*y1
yy2=c*x1+b*y1

g1=p*x2+q*y2
g2=r*x2+s*y2


h1=g1.subs({x2:xx1,y2:yy2})
h2=g2.subs({x2:xx1,y2:yy2})
expand(h1).collect(x1).collect(y1)
```

```python
# http://docs.sympy.org/latest/modules/vector/intro.html
from sympy.vector import CoordSys3D
N=CoordSys3D('N')

u1,u2,u3=symbols('u1 u2 u3')
v1,v2,v3=symbols('v1 v2 v3')
U=u1*N.i+u2*N.j+u3*N.k
V=v1*N.i+v2*N.j+v3*N.k
ex=expand(U.cross(V).dot(U.cross(V)))
```

```python
# |U x V| = |U| |V| -|UV|
expand(U.dot(U)*V.dot(V)-U.dot(V)**2)-ex
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(subplot_kw={'projection':'3d'})

# 起点和终点
start = [3, -2, 4]
end = [5, 6, -2]

# 向量分量 = 终点 - 起点
vector = [end[i] - start[i] for i in range(3)]

ax.quiver(
    start[0], start[1], start[2],       # 起点
    vector[0], vector[1], vector[2],    # 向量分量
    color='blue', linewidth=1.5, arrow_length_ratio=0.1
)
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.set_zlim(-10,10)

import numpy as np
t=np.linspace(-2,1,2)
print(t,3+2*t,-2+8*t,4-6*t)
ax.plot(3+2*t,-2+8*t,4-6*t)
ax.scatter(3,-2,4)
ax.scatter(5,6,-2)

ax.xaxis.pane.set_edgecolor('black')  # 设置坐标平面边框颜色
ax.yaxis.pane.set_edgecolor('black')
ax.zaxis.pane.set_edgecolor('black')
```
